### Capture Techniques

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar os dados do Excel
df = pd.read_excel("/content/capture_technique_count.xlsx")

# 2. Ordenar os dados de forma decrescente pela coluna "Valor"
df = df.sort_values("Count", ascending=False)

# 3. Criar o gráfico de barras horizontal usando Seaborn
plt.figure(figsize=(12, 10))
sns.barplot(x="Count", y="Capture Technique", data=df, palette="magma")


#Ajustar o tamanho da fonte
plt.rc('font', size=14)  # Define o tamanho padrão da fonte para 14
#plt.rc('axes', titlesize=16)  # Tamanho da fonte do título do gráfico
plt.rc('axes', labelsize=14)  # Tamanho da fonte dos rótulos dos eixos
plt.rc('xtick', labelsize=14)  # Tamanho da fonte dos valores do eixo X
plt.rc('ytick', labelsize=20)  # Tamanho da fonte dos valores do eixo Y
plt.rc('legend', fontsize=14)  # Tamanho da fonte da legenda


#plt.title("Gráfico de Barras Horizontal Ordenado")
plt.xlabel("Count")
plt.ylabel("Capture Techniques")
plt.show()



```
